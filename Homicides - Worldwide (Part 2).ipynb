**Clustering Data**

```python
#Import ski-learn libraries for machine learning algorithms
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
features = ggi[['GDP','Gini','IHR']]
```

```python
kmeans.fit(features.dropna())
```

```python
kmeans.labels_
```

```python
g = plt.scatter(x=features.dropna()['IHR'],y=features.dropna()['GDP'],c=kmeans.labels_)
g.axes.set_xlim(0,150)
g.axes.set_ylim(0,60000)
g.axes.set_title('Clustering of variables IHR and GDP')
```

Data shows countries can be classified in four groups where lower intentional homicides rates (per 100.000 people) can be seen in countries where GDP per cápita is higher. It makes sense to say that losses of lives impact economy and viceversa.

```python
g = plt.scatter(x=features.dropna()['Gini'],y=features.dropna()['GDP'],c=kmeans.labels_)
g.axes.set_title('Clustering of variables Gini and GDP')
```

On the other hand, these four groups above are shown when plotting Gini vs GDP. Countries with relatively low Gini indicator have also low GDP per cápita, but it seems not to be a direct relationship, since data also shows that countries with medium and high GDP have considerably high Gini values (more than 25)

```python
g = plt.scatter(x=features.dropna()['IHR'],y=features.dropna()['Gini'],c=kmeans.labels_)
g.axes.set_title('Clustering of variables IHR and Gini')
```

Highest intentional homicides rate appears to be in countries with high Gini values -which means more inequality. However, there are also countries with low homicides rates and high inequality

# Map with Folium Library

The idea is now create a world map with the folium library for the intentional homicidal rates in 2015

```python
import folium
import os
```

```python
world_geo = os.path.join('world-countries.json')
```

```python
# Initialize the map:
m = folium.Map()

# Filter data for 2015
data = ihr[ihr['Year']=='2015'].copy()

# Add the color for the chloropleth:
m.choropleth(
 geo_data=world_geo,
 name='choropleth',
 data=data,
 columns=['Country Code', 'IHR'],
 key_on='feature.id',
 fill_color='YlGn',
 fill_opacity=0.7,
 line_opacity=0.2,
 legend_name='IHR per 100.000 people'
)
folium.LayerControl().add_to(m)
 
# Save to html
m.save('#1_folium_chloropleth_IHR_World_2015.html')
```
